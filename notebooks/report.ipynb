# Uber Stock (UBER) — Financial Returns & Risk Report

This report analyzes Uber’s stock using *adjusted prices and daily returns* (not raw price levels).  
The workflow focuses on return behavior, volatility clustering, drawdowns, and risk metrics (annualized return/volatility, Sharpe, VaR/CVaR), and includes a simple baseline model to predict next-day returns using lagged returns and volume changes.

**Key interpretation notes**
- Daily returns are typically closer to stationary than prices and are the standard unit for risk modeling.
- Large drawdowns highlight downside risk that can be hidden in price-only charts.
- Predicting next-day returns is difficult in liquid markets; low/negative R² is common for simple baselines.
